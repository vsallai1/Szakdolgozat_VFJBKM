```python
df['difficulty_average'] = pd.to_numeric(df['difficulty_average'], errors='coerce')
df['downloadable']=df['downloadable'].astype(bool)
df['free']=df['free'].astype(bool)
df['yarn_weight_id']=df['yarn_weight_id'].astype('int64')
df['rating_average'] = pd.to_numeric(df['rating_average'], errors='coerce')
df['needle_metric'] = pd.to_numeric(df['needle_metric'], errors='coerce')
df["pattern_creation_date"]=pd.to_datetime(df['pattern_creation_date'],utc=True,format='mixed').dt.tz_localize(None)
df['pattern_type_id']=df['pattern_type_id'].astype('int64')
df['pattern_source_type_id']=df['pattern_source_type_id'].astype('int64')
df['yarn_id']=df['yarn_id'].astype('int64')
df['yarn_company_id']=df['yarn_company_id'].astype('int64')
df.dtypes
```
